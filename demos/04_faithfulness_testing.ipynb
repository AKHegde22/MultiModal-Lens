# Demo 04: Faithfulness Diagnostics

Run deletion, insertion, counterfactual drop, and attention-gradient Spearman rank correlation tests.

```python
from multimodallens import LensPipeline
from PIL import Image
import numpy as np

pipe = LensPipeline()
img = Image.fromarray(np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8))

result = pipe.analyze(model_name="openai/clip-vit-base-patch32", image=img, prompt="a cat", run_faithfulness=True)

if result.faithfulness:
    print("Counterfactual Drop:", result.faithfulness.counterfactual_drop)
    print("Attn-Grad Spearman:", result.faithfulness.attn_grad_spearman)
```
