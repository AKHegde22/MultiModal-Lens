# Demo 01: Quickstart Guide

This tutorial shows how to analyze a vision-language model in under 5 lines of Python code using `HookedVLM`.

```python
from multimodallens import HookedVLM
from PIL import Image
import numpy as np

# 1. Load model
model = HookedVLM.from_pretrained("openai/clip-vit-base-patch32")

# 2. Analyze image and prompt
image = Image.fromarray(np.zeros((224, 224, 3), dtype=np.uint8))
result = model.analyze(image, "a photo of a cat")

print(f"Global Score: {result.global_score}")
print(f"Extracted Tokens: {result.tokens}")
```
