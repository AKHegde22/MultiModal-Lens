# Demo 03: Mechanistic Probes

Explore logit lens, activation patching, and grounding head discovery.

```python
from multimodallens import LensPipeline
from PIL import Image
import numpy as np

pipe = LensPipeline()
img = Image.fromarray(np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8))

# 1. Logit Lens
lens = pipe.logit_lens(model_name="openai/clip-vit-base-patch32", image=img, prompt="a dog")
print(f"Logit lens decoded {len(lens.steps)} layer steps.")

# 2. Grounding Heads
grounding = pipe.grounding_heads(model_name="openai/clip-vit-base-patch32", image=img, prompt="a dog")
print(f"Ranked {len(grounding.heads)} candidate grounding heads.")
```
