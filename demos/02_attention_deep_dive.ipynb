# Demo 02: Attention Rollout & Token Alignment

Deep dive into attention maps, token-patch similarity matrices, and gradient attribution.

```python
from multimodallens import LensPipeline
from PIL import Image
import numpy as np

pipe = LensPipeline()
image = Image.fromarray(np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8))

# Analyze with gradient attributions enabled
res = pipe.analyze(model_name="openai/clip-vit-base-patch32", image=image, prompt="a red bus", compute_gradients=True)

print("Available attention maps:", list(res.attention_maps.keys()))
print("Alignment matrix shape:", res.alignment_matrix.shape)
```
