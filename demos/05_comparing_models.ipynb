# Demo 05: Comparing Prompts & Models

Compare prompt sensitivities side-by-side.

```python
from multimodallens import LensPipeline
from PIL import Image
import numpy as np

pipe = LensPipeline()
img = Image.fromarray(np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8))

res_a, res_b = pipe.compare_prompts(
    model_name="openai/clip-vit-base-patch32",
    image=img,
    prompt_a="a photo of a cat",
    prompt_b="a photo of a car"
)

print(f"Prompt A Score: {res_a.global_score:.4f}")
print(f"Prompt B Score: {res_b.global_score:.4f}")
print(f"Delta (A - B):  {res_a.global_score - res_b.global_score:.4f}")
```
