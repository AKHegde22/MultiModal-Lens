# MultimodalLens: Quickstart & Features Demo

Welcome to **MultimodalLens** — a library for mechanistic interpretability and debugging of vision-language models (VLMs), inspired by TransformerLens.

This demo walks through:
1. Loading a VLM with `HookedVLM`
2. Extracting attention rollout and token-image alignment
3. Capturing internal layer activations with `run_with_cache`
4. Performing interventions with `run_with_hooks`
5. Decoding intermediate layer predictions with Logit Lens

```python
# 1. Install & Imports
!pip install -q multimodallens pillow matplotlib torch

from multimodallens import HookedVLM, LensPipeline
from PIL import Image
    import numpy as np
import matplotlib.pyplot as plt

print("MultimodalLens imported successfully!")
```

## 2. Load Model via `HookedVLM`

Just like TransformerLens's `HookedTransformer`, `HookedVLM.from_pretrained()` loads any vision-language model checkpoint in 1 line.

```python
# Load model
vlm = HookedVLM.from_pretrained("openai/clip-vit-base-patch32", device="auto")
print(f"Loaded model: {vlm.model_name} (family: {vlm.family})")
print(f"Device: {vlm.device}")
```

## 3. Basic Analysis & Visualizations

```python
# Create sample synthetic image (or open a real JPEG/PNG)
img_arr = np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8)
image = Image.fromarray(img_arr)

result = vlm.analyze(image=image, prompt="a photo of a cat")

print(f"Global Score: {result.global_score:.4f}")
print(f"Tokens: {result.tokens}")
print(f"Patch Grid: {result.patch_grid}")
```

## 4. `run_with_cache`: Access Internal Activations

`vlm.run_with_cache()` executes the forward pass and returns an `ActivationCache` containing named layer tensors.

```python
# List available hook points in the model
hook_points = vlm.list_hook_points()
print(f"Discovered {len(hook_points)} hook points. First 5:")
for hp in hook_points[:5]:
    print(f"  - {hp}")

# Run forward pass with caching
result, cache = vlm.run_with_cache(image, "a photo of a dog")

# Access activations dict-style by layer name
layer_0_act = cache[hook_points[0]]
print(f"Layer 0 activation shape: {layer_0_act.shape}")
```

## 5. `run_with_hooks`: Interventions & Patching

Intervene on model computation on the fly by passing forward hook functions.

```python
# Define zero ablation hook
def zero_ablate_layer(tensor):
    return tensor * 0.0

# Run model under intervention
patched_result = vlm.run_with_hooks(
    image=image,
    prompt="a photo of a cat",
    fwd_hooks=[(hook_points[0], zero_ablate_layer)]
)

print(f"Baseline Global Score: {result.global_score:.4f}")
print(f"Patched  Global Score: {patched_result.global_score:.4f}")
```

## 6. High-Level LensPipeline API

For multi-prompt comparison, logit lens, or batch evaluation, use `LensPipeline`:

```python
pipe = LensPipeline()

# Logit lens decoding across intermediate layers
lens_res = pipe.logit_lens(model_name="openai/clip-vit-base-patch32", image=image, prompt="a cat")
print(f"Logit lens decoded {len(lens_res.steps)} steps.")
if lens_res.steps:
    first = lens_res.steps[0]
    print(f"Layer {first.layer_name} top tokens: {first.top_tokens}")
```
